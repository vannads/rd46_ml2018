## 1. Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = np.random.uniform(1, 100, 1000000000)
plt.hist(a)
```

## 2. Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = np.random.randn(100)
b = range(1, 101)

plt.plot(b,a)
```

Standard Normal Distrubution :
A normal distribution with mean=0, and SD=1 is called a Standard Normal Distrubution
